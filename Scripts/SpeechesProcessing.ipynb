##Data Processing dei dati testuali relativi ai discorsi pronunciati da un politico contenuti in una cartella

```python
from os import listdir
```

```python
#Visualizzazione dei nomi dei file
nomi_file = listdir("/content/drive/MyDrive/Speeches")
nomi_file
```

```python
import pandas as pd
```

```python
#Creazione di un Datframe che conterr√† i miei dati
df = pd.DataFrame(index=range(len(nomi_file)), columns=['Nome_del_file', 'Contenuto'])
df.head()
```

```python
#Ciclo for che mi permette di aprire ogni file e copiare suo contenuto nella colonna "Contenuto"

for i, f_name in enumerate(nomi_file):
    f = open("/content/drive/MyDrive/Speeches/" + f_name, 'r', encoding="utf-8")
    f_content = f.readlines()
    f.close()

    df.at[i, "Nome_del_file"] = f_name
    df.at[i, "Contenuto"] = f_content[0]
```

```python
df
```

```python
#Creazione di una lista contenente le abbreviazioni dei mesi come riportati nella colonna nome del file
mesi = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
mesi
```

```python
#Funzione che ci permette di recuperare il nome della citt√†
def SeparaCitta(c):
  for m in mesi:
    if (m in c):
      citta = c[:c.find(m)]
  return citta
```

```python
#Creazione di una colonna Citt√†
df['Citta'] = df.Nome_del_file.apply(SeparaCitta)
df
```

```python
#Estrazione della data
def SeparaData(c):
  date = c.Nome_del_file[len(c.Citta):c.Nome_del_file.find(".txt")]
  return date
```

```python
df['Data'] = df.apply(SeparaData, axis=1)
df
```

```python
df.dtypes
```

```python
#Trasformazione della data in formato datetime
df['Data'] = pd.to_datetime(df['Data'], format="%b%d_%Y")
df.head()
```

```python
#Cancellazione della colonna Nome_del_file
df.drop('Nome_del_file', axis=1, inplace = True)
```

```python
df.head()
```

```python
#Estrazione delle informazioni relativi alla data

#Anno
df["Anno"] = df.Data.dt.year
#Mese
df['Mese'] = df.Data.dt.month
#Giorno del mese
df['Giorno_mese'] = df.Data.dt.day
#Giorno della settimana
df['Giorno_settimana'] = df.Data.dt.weekday
```

```python
df.head()
```

```python
import numpy as np
```

```python
#Capire se un giorno √® un weekend o no
df['Weekend'] = np.where((df.Giorno_settimana == 5) | (df.Giorno_settimana == 6), "S√¨", "No")
```

```python
#Salvo il file in un formato csv
df.to_csv("cleaned-speeches.csv", index=False)
```
