Numero di disoccupati (valore "Unemployed" del campo "Employment Status") tra le persone con un grado di studio (campo "Education Level") pari alle scuole superiori (valore "High School").

**Interrogazione 5**

Per ogni stato di impiego (campo "Employment Status"), mostra il numero di persone in quello stato, ed il loro guadagno settimanale (campo "Weekly Earnings") medio e massimo.

```python
pipeline = [
    {'$project': {                                          # riduco attributi per performance
        'Employment Status':1,
        'Weekly Earnings':1
    }},
    {'$group' : {
        '_id' : '$Employment Status',
        'numero di persone con stato' : {'$count':{}},      # 1 persona 1 documento, se intervistata in piu anni il suo guadagno fara media
        'guadagno settimanale max' : {'$max':'$Weekly Earnings'},
        'guadagno settimanale medio' : {'$avg':'$Weekly Earnings'}
    }},
    {'$project' : {
        '_id':0,
        'Stato occupazionale':'$_id',
        'numero di persone con stato':1,
        'guadagno settimanale max':1,
        'guadagno settimanale medio':1
    }}
]

util(pipeline)
```

**Interrogazione 6**

Il numero e l'elenco delle persone che hanno compilato il questionario tra il 2005 e il 2008 e che sono disoccupate, divise per fascia d'età (campo "Age Range"): il risultato deve essere ordinato per numero di persone in ogni fascia ordinate in ordine crescente.

```python
pipeline = [
    {'$project' : {                                           # performance
        'Year':1,
        'Employment Status':1,
        'Age Range':1
    }},
    {'$match' : { 'Year' : { '$gte' : 2005,'$lte' : 2008 }}}, # si è supposto COMPRESI
    {'$match' : { 'Employment Status' : 'Unemployed' }},
    {'$group' : {
        '_id' : '$Age Range',
        'conteggio' : { '$count' : {}},
        'elenco persone' : { '$push' : '$_id' } # 1 persona 1 documento 1 intervista in determinato anno quindi push
    }},
    {'$sort' : { 'conteggio' : 1}},
    {'$project' : {
        '_id':0
    }}
]

util(pipeline)
```

## **Machine Learning e Deep Learning con Python**
L'American Time Use Survey raccoglie dati statistici riguardanti le abitudini dei cittadini americani. Lo scopo di questa prova è implementare e testare le performance di un classificatore che sia in grado di predire lo stato occupazionale dei cittadini in base alle loro abitudini.

### Acquisizione e pre-processamento del dataset
1. Importare il dataset 'WA_American-Time-Use-Survey-lite.csv' in un Pandas DataFrame


```python
data_df = pd.read_csv("/content/drive/MyDrive/PROVA_FINALE/WA_American-Time-Use-Survey-lite.csv")
data_df
```

2. Selezionare le colonne corrispondenti alle abitudini 'Sleeping', 'Shopping', 'Eating and Drinking', 'Socializing & Relaxing','Television', 'Golfing', 'Running', 'Volunteering' e salvarle in una variabile X. Selezionare la colonna "Employment" e salvarla nella variabile y_raw

```python
data_df.columns
```

```python
X_columns = ['Sleeping', 'Shopping', 'Eating and Drinking',\
             'Socializing & Relaxing','Television', 'Golfing',\
             'Running', 'Volunteering']
X = data_df[X_columns]

y_raw = data_df['Employment Status'].tolist() # struttura dati pandas.Series -> List
```

3. Convertire i valori di y_raw in valori binari: creare una nuova lista y che contiene 1 quando y_raw contiene il valore 'Employed' e 0 quando contiene un altro valore

```python
y = [1 if status == 'Employed' else 0 for status in y_raw]
```

4. Dividere i dati in train (80% dei dati) e test set (20% dei dati), creando le variabili X_train, X_test, y_train, y_test

```python
# determinismo del modello settando un valore per il paramentro random_state
random_state = 42
```

```python
# partizionamento in train 80% e test set 20%
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=random_state, shuffle=True)
```

### Implementare una regressione logistica e studiare la precision, recall e F1
1. Definisci il modello LogisticRegression creando una variabile model_log


```python
# selezione dell'algoritmo
from sklearn.linear_model import LogisticRegression

model_log = LogisticRegression(random_state=random_state)
```

2. Addestra il modello sul training set


```python
# addastramento su training set SUPERVISIONATO : dati+label
model_log.fit(X_train,y_train)
```

3. Calcola le predizioni del modello sul test set, salvandole in una variabile predicted_log

```python
# previsione su TEST-SET del modello attuale
predicted_log = model_log.predict(X_test) # classi predette salvate per confronto successivo con quelle note
```

4. Calcola la precisione p_log, la recall r_log, e la F1 f1_log del classificatore sul test set

```python
# calcolo delle metriche sul modello attuale (SENZA VALIDATION) rispetto alle previsioni su TEST-SET
from sklearn.metrics import precision_score, recall_score, f1_score

p_log = precision_score(predicted_log, y_test)
r_log = recall_score(predicted_log, y_test)
f1_log = f1_score(predicted_log, y_test)
```

```python
print(f'Metriche su: {model_log.__class__.__name__}\n' + \
      f'precisione: {p_log}\n' + \
      f'recall: {r_log}\n' + \
      f'F1 score: {f1_log}')
```

5. Calcola la F1 f1_cv_log del modello usando la 10-fold cross validation

```python
# implementazione delle cross validation su LogisticRegression
from sklearn.model_selection import KFold, cross_validate
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score

# numero di split della cv
n_splits = 10

kfold = KFold(
    n_splits=n_splits,
    random_state=random_state,
    shuffle=True)

# classificatore
model_log_cv = LogisticRegression(random_state=random_state)

# funzione di scoring da utilizzare per la validazione
scoring = { 'f1_cv_log': make_scorer(f1_score) }

# cross validation del modello
results_mod_log_cv = cross_validate(model_log, X, y, cv=kfold, scoring=scoring)

results_mod_log_cv
```

### Implementare un percettrone e studiare la precision, recall e F1
1. Definisci il modello Perceptron creando una variabile model_lin


```python
# selezione dell'algoritmo
from sklearn.linear_model import Perceptron

model_lin = Perceptron(random_state=random_state)
```

2. Addestra il modello sul training set


```python
# addastramento su training set SUPERVISIONATO : dati+label
model_lin.fit(X_train,y_train)
```

3. Calcola le predizioni del modello sul test set, salvandole in una variabile predicted_lin

```python
# previsione su TEST-SET del modello attuale
predicted_lin = model_lin.predict(X_test)
```

4. Calcola la precisione, recall e F1

```python
# calcolo delle metriche sul modello attuale (SENZA VALIDATION) rispetto alle previsioni su TEST-SET

p_lin = precision_score(predicted_lin, y_test)
r_lin = recall_score(predicted_lin, y_test)
f1_lin = f1_score(predicted_lin, y_test)
```

```python
print(f'Metriche su: {model_lin.__class__.__name__}\n' + \
      f'precisione: {p_lin}\n' + \
      f'recall: {r_lin}\n' + \
      f'F1 score: {f1_lin}')

# da notare che il modello sviluppato a partire dal Perceptron ha una
# precision molto minore del corrispodente sviluppato con LogisticRegression
# per questo motivo la media armonica F1 score è minore in quanto "sensibile" ai valori piccoli
```

5. Calcola la F1 f1_cv_lin del modello usando la 10-fold cross validation

```python
# implementazione delle cross validation su Perceptron

# numero di split della cv
n_splits = 10

kfold = KFold(
    n_splits=n_splits,
    random_state=random_state,
    shuffle=True)

# classificatore
model_lin_cv = Perceptron(random_state=random_state)

# funzione di scoring da utilizzare per la validazione
scoring = { 'f1_cv_lin': make_scorer(f1_score) }

# cross validation del modello
results_mod_lin_cv = cross_validate(model_lin, X, y, cv=kfold, scoring=scoring)

results_mod_lin_cv
```

1. Stampare il nome del classificatore che ha la F1 piu' alta in base calcolata dalla cross-validation

```python
import numpy as np

max_F1_model_log_cv = np.amax(results_mod_log_cv['test_f1_cv_log'])
max_F1_model_lin_cv = np.amax(results_mod_lin_cv['test_f1_cv_lin'])

greatest_F1_model = ''

if max_F1_model_lin_cv >= max_F1_model_log_cv:          # se uguali considero il primo
    greatest_F1_model = model_lin_cv.__class__.__name__
else:
    greatest_F1_model = model_log_cv.__class__.__name__

print(f'Best model: {greatest_F1_model}')
```
