Lo scopo di questa prova è implementare e testare le prestazioni di un classificatore che sia in grado di predire lo stato occupazionale dei cittadini in base alle loro abitudini.

1) Importare il dataset `'WA_American-Time-Use-Survey-lite.csv'` in un Pandas DataFrame [Peso: 0.075]

````python
import pandas as pd

filename = "WA_American-Time-Use-Survey-lite.csv"

data = pd.read_csv(filename)
data
````

2) Selezionare le colonne corrispondenti alle abitudini `['Sleeping', 'Shopping', 'Eating and Drinking', 'Socializing & Relaxing','Television', 'Golfing', 'Running', 'Volunteering']` e salvarle in una variabile `X`. Selezionare la colonna `"Employment Status"` e salvarla nella variabile `y_raw` [Peso: 0.075]

````python
import numpy as np

abitudini = ["Sleeping", "Shopping", "Eating and Drinking", "Socializing & Relaxing",
         "Television", "Golfing", "Running", "Volunteering"]

X = data[abitudini]
y_raw = data["Employment Status"]
````

3) Convertire i valori di `y_raw` in valori **binari**: creare una nuova lista `y` che contiene 1 quando `y_raw` contiene il valore `'Employed'` e 0 quando contiene un altro valore [Peso: 0.075]


````python
y = []

for line in y_raw:
    if line == 'Employed':
        y.append(1)
    else: 
        y.append(0)
````

4) Dividere i dati in train (80% dei dati) e test set (20% dei dati), creando le variabili `X_train`, `X_test`, `y_train`, `y_test` [Peso: 0.075]

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
````

5) Implementare il modello SVM lineare e fare il fit sugli esempi del **training set** [Peso: 0.2]

````python
from sklearn.svm import LinearSVC

svm = LinearSVC()
svm.fit(X_train,y_train)
````

6) Calcolare le predizioni `y_prediction` sul test set [Peso: 0.1]

````python
y_prediction = svm.predict(X_test)
````

7) Misurare accuracy, precisione, recall, F1 [Peso: 0.2]

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

a = accuracy_score(y_prediction, y_test)
p = precision_score(y_prediction, y_test)
r = recall_score(y_prediction, y_test)
f1 = f1_score(y_prediction, y_test)

print("Accuracy:", a)
print("Precision:", p)
print("Recall:", r)
print('F1:', f1)
````

8) Implementare una 5-fold cross validation e misurare l'accuracy, precision, recall e F1 [Peso: 0.2]

````python
from sklearn.model_selection import KFold, cross_validate
from sklearn.metrics import make_scorer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

kfold = KFold(n_splits=5, random_state=7, shuffle=True)
svm_cv = LinearSVC()

scoring = {'acc': make_scorer(accuracy_score), 'p': make_scorer(precision_score), 'r':make_scorer(recall_score) ,'f1' : make_scorer(f1_score)}

results = cross_validate(svm_cv, X, y, cv=kfold, scoring=scoring)
#print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
results
````

Lo scopo di questa prova è implementare e testare le prestazioni di un classificatore basato su rete neurale che sia in grado di classificare correttamente il numero all'interno dell'immagine.

Il seguente blocco di codice scarica il dataset e importa le principali librerie necessarie per questo esercizio. La variabile `x` contiene un `numpy.ndarray` contente i vettori di pixels delle immagini (su 1 dimensione). La variabile `y` contiene un `numpy.ndarray` contenente, per ogni immagine, l'annotazione corrispondente (i.e., il numero all'interno dell'immagine da 0 a 9 inclusi).Il seguente blocco di codice scarica il dataset e importa le principali librerie necessarie per questo esercizio. La variabile `x` contiene un `numpy.ndarray` contente i vettori di pixels delle immagini (su 1 dimensione). La variabile `y` contiene un `numpy.ndarray` contenente, per ogni immagine, l'annotazione corrispondente (i.e., il numero all'interno dell'immagine da 0 a 9 inclusi).

````python
#NON TOCCARE

! pip install torch scikit-learn
import torch
import torch.nn as nn
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', cache=False)

x = mnist.data.astype('float32') #immagini
y = mnist.target.astype('int64') #labels (0, 1, ..., 9)
````

1) Trasformare `x` e `y` in **tensori** utilizzando la libreria PyTorch. Stampare inoltre la dimensione dei due tensori risultanti. \[Peso=0.15\]

````python
x = torch.tensor(x.to_numpy())
y = torch.tensor(y.to_numpy())

print('x shape: {}'.format(x.shape))
print('y shape: {}'.format(y.shape))
````

2) \[Questo esercizio non è necessario per il proseguimento\] Estrarre il primo esempio e la corrispondente annotazione. Convertire il vettore immagine in una matrice 28x28. Posizionare i risultati rispettivamente nelle variabili `img` e `label`. Se l'operazione è stata effettuate con successo, il prossimo blocco di codice stamperà a video il primo esempio del dataset (che contiene un 5). \[Peso=0.05\]

````python
img   = x[0].view(28, 28)
label = y[0]
````

Il seguente blocco contiene il codice per stampare a video la prima immagine del dataset e la sua annotazione.

````python
#NON TOCCARE

from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms as T
from IPython.display import display

print('## Primo esempio del dataset')
plt.imshow(Image.fromarray(img.numpy()), cmap='gray')
print('Label: {}'.format(label))

x /= 255.0 #normalizzazione
````

3) Dividere il dataset nei seguenti split: training (80%), validation (10%) e test (10%)\[Peso=0.05\]

````python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.1, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=.1, random_state=42)
````

4) Creare la struttura `Dataset` per MNIST. Instanziare inoltre il `DataLoader` per gli split di training e validation utilizzando la libreria PyTorch. Usare una dimensione di `batch=32` ed attivare lo `shuffle` durante il training.\[Peso=0.15\]

````python
from torch.utils.data import Dataset, DataLoader


class MNISTDataset(Dataset):
  def __init__(self, images, labels):
    self.images = x
    self.labels = y

  def __getitem__(self, index):
    return self.images[index], self.labels[index]

  def __len__(self):
    return len(self.images)

device       = torch.device('cuda:0' if torch.cuda.is_available() and False else 'cpu')
train_data   = MNISTDataset(images=x_train, labels=y_train)
val_data     = MNISTDataset(images=x_val, labels=y_val)

train_loader = DataLoader(train_data,
                          batch_size=32,
                          shuffle=True,
                          num_workers=2,
                          pin_memory=(device.type == 'cuda'))
val_loader   = DataLoader(val_data,
                          batch_size=32,
                          shuffle=False,
                          num_workers=2,
                          pin_memory=(device.type == 'cuda'))
````

5) Creare una rete neurale **Fully Connected** che contenga 1 hidden layer seguito da una non linearità.
Inoltre, istanziare un ottimizzatore (Adam) con `lr=1e-3` e definire una funzione di loss (Cross Entropy) utilizzando la libreria PyTorch.
Stampare il modello così definito. \[Peso=0.1\]

````python
from torch.optim import Adam
from torch.nn import CrossEntropyLoss

class MyNet(nn.Module):
  def __init__(self):
    super(MyNet, self).__init__()
    self.net = nn.Sequential(nn.Linear(784, 784//2),
                              nn.Linear(784//2, 784//4),
                              nn.ReLU(),
                              nn.Linear(784//4, 10))
  
  def forward(self, img):
    return self.net(img)

model     = MyNet().to(device)
optimizer = Adam(model.parameters(), lr=1e-3)
criterion = CrossEntropyLoss().to(device)
print(model)
````

6) Implementare un ciclo di allenamento del modello. Tale ciclo deve essere effettuato per 3 epoche, al termine di ognuna delle quali è necessario stampare a video le informazioni riguardo la loss media per training e la loss media su validation (es. 'Epoch #5, train_loss=0.53, val_loss=0.13').
Infine, salvare su disco il modello che ottiene i risultati di loss migliori. \[Peso=0.25\]


````python
# circa 20s per epoca su cpu

from tqdm import tqdm
import math

num_epochs = 3
t_losses   = []
v_losses   = []
best_loss  = math.inf

for ep in tqdm(range(1, num_epochs+1)):
  curr_losses = []
  model.train()
  for imgs, labels in train_loader:
    imgs   = imgs.to(device)
    labels = labels.to(device)
    
    optimizer.zero_grad()

    logits = model(imgs)
    loss   = criterion(logits, labels)
    loss.backward()
    optimizer.step()
    curr_losses.append(loss.item())
  t_losses.append(sum(curr_losses) / len(curr_losses))
  curr_losses = []
  model.eval()
  with torch.no_grad():
    for imgs, labels in val_loader:
      imgs   = imgs.to(device)
      labels = labels.to(device)
      logits = model(imgs)
      loss = criterion(logits, labels)
      curr_losses.append(loss.item())
  v_losses.append(sum(curr_losses) / len(curr_losses))
  if v_losses[-1] < best_loss:
    best_loss = v_losses[-1]
    torch.save(model, 'best_model.pth')
  print('Epoch #{}: train_loss={}, val_loss={}'.format(ep, t_losses[-1], v_losses[-1]))
````

7) Una volta che il modello più performante è stato salvato, caricarlo in una variabile. Dopodichè, effettuare il calcolo delle seguenti metriche sull'insieme dei dati di test, facendone una stampa a video:
- Precision
- Recall
- F1

> Nota: durante lo score delle metriche (es. accuracy), dato che il dataset è multiclasse bisogna utilizzare il parametro `average='micro'`

\[Peso=0.25\]

````python
import torch.nn.functional as F
from sklearn.metrics import precision_score, recall_score, f1_score

model = torch.load('best_model.pth', map_location=device)

test_data = MNISTDataset(images=x_test, labels=y_test)
test_loader   = DataLoader(test_data,
                           batch_size=32,
                           shuffle=False,
                           num_workers=2,
                           pin_memory=True)

true_labels = []
preds       = []
model.eval()
with torch.no_grad():
  for img, label in test_loader:
    img        = img.to(device)
    labels     = labels.to(device)
    logits     = model(img)
    curr_preds = torch.argmax(F.softmax(logits, dim=-1), dim=-1)
    true_labels.extend(label.view(-1).tolist())
    preds.extend(curr_preds.view(-1).tolist())

#todo precision, recall, f1
print('Precision: {}'.format(precision_score(true_labels, preds, average='micro')))
print('Recall: {}'.format(recall_score(true_labels, preds, average='micro')))
print('F1: {}'.format(f1_score(true_labels, preds, average='micro')))
````
